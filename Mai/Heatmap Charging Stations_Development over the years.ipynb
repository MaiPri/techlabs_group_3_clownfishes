```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

# DE Karte
germany = gpd.read_file('plz-1stellig.shp')


# read csv
charging_data = pd.read_csv('charging_data.csv')

# Inbetriebnahmedatums in Datetime konvertrieren
charging_data['inbetriebnahmedatum'] = pd.to_datetime(charging_data['inbetriebnahmedatum'])

# Filter Inbetriebnahme vor oder in 2015
charging_data_before_2015 = charging_data[charging_data['inbetriebnahmedatum'].dt.year <= 2010]

# Geo Data Frame
charging_gdf = gpd.GeoDataFrame(charging_data_before_2015, geometry=gpd.points_from_xy(charging_data_before_2015.laengengrad, charging_data_before_2015.breitengrad))

# DE Karte
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
germany.plot(ax=ax, color='lightgrey', edgecolor='black')

# Kadestationen = rote Punkte
charging_gdf.plot(ax=ax, color='red', markersize=5, alpha=0.5)

# Top 10 Stdte Koordinaten
cities = {
    'Berlin': (13.4049, 52.5200),
    'Hamburg': (9.9937, 53.5511),
    'Munich': (11.5819, 48.1351),
    'Cologne': (6.9603, 50.9375),
    'Frankfurt': (8.6821, 50.1109),
    'Stuttgart': (9.1829, 48.7758),
    'Dsseldorf': (6.7735, 51.2277),
    'Dortmund': (7.4686, 51.5136),
    'Essen': (7.0117, 51.4556),
    'Leipzig': (12.3874, 51.3438)
}

# Top 10 Stdte auf KArte einzeichnen
for city, coords in cities.items():
    ax.plot(coords[0], coords[1], marker='o', color='black', markersize=10)
    ax.text(coords[0], coords[1], city, fontsize=10, ha='right', va='bottom')

# show map
plt.title('Charging Stations in Germany (Before or in 2010)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
